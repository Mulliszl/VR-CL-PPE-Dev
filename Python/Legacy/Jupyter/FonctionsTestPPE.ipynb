```python
from numpy import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = linspace(1, 20,500)
xp = linspace(3,9,500)
x1 = 0.45*xp**2 + -10*xp + 50


fct = -log(x)+8
rfct = ndarray.round(fct)
fcta = -x/8+7
plt.figure(figsize=(16, 6), dpi=80)
plt.ylim(0, 10)
plt.xlim(0, 20)
plt.plot(x,fct,label='logarithmique')
#plt.plot(x,rfct)
plt.plot(x,fcta,label='continue')
plt.plot(x1,xp,label='polynomiale')

plt.legend()
```

```python
x = linspace(1, 100,500)
xp = linspace(0,9,500)
x1 = 0.45*xp**2 + -10*xp + 50
fct = -log(x)+8
rfct = ndarray.round(fct)
fcta = -x/8+7
plt.figure(figsize=(16, 6), dpi=80)
plt.ylim(0, 10)
plt.xlim(0, 100)
plt.plot(x,fct,label='logarithmique')
#plt.plot(x,rfct)
plt.plot(x,fcta,label='continue')
plt.plot(x1,xp,label='polynomiale')
plt.legend()
```

```python
x = linspace(1, 100,500)
plt.figure(figsize=(16, 6), dpi=80)
plt.ylim(0, 8)
fct = -log(x)+8
rfct = ndarray.round(fct)
plt.xlabel("iterations")
plt.ylabel("Allowed time to respond")
plt.plot(x,fct,label='ideal')
plt.plot(x,rfct,label='real')
plt.legend()
```
